## Step 1: Import Required Libraries
We import necessary libraries for data handling, NLP preprocessing, and machine learning.

```python
import pandas as pd
import numpy as np
import re
import string

# NLP Libraries
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# ML Libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

## Step 2: Load Dataset

We use a sentiment dataset (e.g., IMDb reviews) containing text reviews and sentiment labels.

```python
df = pd.read_csv("IMDB Dataset.csv")

# Display first 5 rows
df.head()

# Dataset info
print("Total Samples:", len(df))
print(df['sentiment'].value_counts())
```

## Step 3: NLP Preprocessing

We perform the following preprocessing steps:
- Convert text to lowercase
- Remove punctuation and numbers
- Remove stopwords
- Tokenization
- Lemmatization
- Remove URLs and special characters

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    
    #1 Convert to lowercase
    text = text.lower()
    
    #2 Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    #3 Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    #4 Remove numbers
    text = re.sub(r'\d+', '', text)
    
    #5 Tokenization
    tokens = word_tokenize(text)
    
    #6 Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    #7 Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return " ".join(tokens)
```

```python
# Apply preprocessing
df['clean_text'] = df['review'].apply(preprocess_text)

# Show processed data
df[['review', 'clean_text']].head()
```

## Step 4: Feature Engineering

We convert text into numerical form using:
- Bag of Words (BoW)
- TF-IDF

```python
# Bag of Words

bow = CountVectorizer(max_features=5000)
X_bow = bow.fit_transform(df['clean_text']).toarray()
```

```python
# TF-IDF

tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df['clean_text']).toarray()
```

```python
# Target Variable

# Convert labels into numeric
y = df['sentiment'].map({'positive':1, 'negative':0})
```

## Step 5: Train-Test Split

We split the dataset into training and testing sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

## Step 6: Model Training

We train three machine learning models:
- Logistic Regression
- Naive Bayes
- Decision Tree

```python
# Logistic Regression

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
# Naive Bayes

nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
```

```python
# Decision Tree

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
```

## Step 7: Model Evaluation

We evaluate models using:
- Accuracy
- Precision
- Recall
- F1 Score

```python
def evaluate(y_test, y_pred, model_name):
    print(f"\n{model_name} Performance:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:", recall_score(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))

# Evaluate all models
evaluate(y_test, y_pred_lr, "Logistic Regression")
evaluate(y_test, y_pred_nb, "Naive Bayes")
evaluate(y_test, y_pred_dt, "Decision Tree")
```

## Step 8: Model Comparison

We compare the performance of all models.

```python
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Naive Bayes', 'Decision Tree'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr),
        accuracy_score(y_test, y_pred_nb),
        accuracy_score(y_test, y_pred_dt)
    ]
})

print(results)
```

## Conclusion

- NLP preprocessing significantly improves text quality.
- TF-IDF performs better than Bag of Words in most cases.
- Logistic Regression achieved the highest accuracy.
- Naive Bayes is faster but slightly less accurate.
- Decision Tree tends to overfit on text data.

Overall, Logistic Regression with TF-IDF is the best combination for this task.
