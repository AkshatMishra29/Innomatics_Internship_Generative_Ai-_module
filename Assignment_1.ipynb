## Task 1: Replace Hardcoded Prompts
Converting a standard function into a LangChain `PromptTemplate`.

```python
from langchain_core.prompts import PromptTemplate

# Task 1: Replace Hardcoded Prompts
template_1 = "Explain {topic} in simple terms for beginners"

prompt_1 = PromptTemplate(
    input_variables=["topic"],
    template=template_1
)

print(prompt_1.format(topic="Quantum Computing"))
```

## Task 2: Multi-Input Prompt System
Creating a flexible template for different topics, audiences, and tones.

```python
from langchain_core.prompts import PromptTemplate

# Task 2: Multi-Input Prompt System
multi_input_template = "Explain {topic} for {audience} in a {tone} tone"

prompt_2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_input_template
)

test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for test in test_cases:
    print(prompt_2.format(**test))
```

## Task 3: Prompt Variations Engine
Generating different prompt styles for the same topic.

```python
from langchain_core.prompts import PromptTemplate

# Task 3: Prompt Variations Engine
teaching_prompt = PromptTemplate.from_template("Teaching → Explain {topic} clearly step by step")
interview_prompt = PromptTemplate.from_template("Interview → Ask 3 questions about {topic}")
story_prompt = PromptTemplate.from_template("Storytelling → Explain {topic} as a story")

topic = "Machine Learning"
print(teaching_prompt.format(topic=topic))
print(interview_prompt.format(topic=topic))
print(story_prompt.format(topic=topic))
```

## Task 4: ChatPromptTemplate System
Using System and User roles to define behavior.

```python
from langchain_core.prompts import ChatPromptTemplate

# Task 4: ChatPromptTemplate System
def get_chat_prompt(role, topic):
    system_template = "You are a helpful {role}."
    human_template = "Explain {topic} to me."

    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_template),
        ("human", human_template),
    ])

    return chat_prompt.format_messages(role=role, topic=topic)

# Example usage
chat_output = get_chat_prompt(role="teacher", topic="Neural Networks")
for msg in chat_output:
    print(msg.type.upper() + ": " + msg.content)
```

## Task 5 & 6: Input Validation and Generator App
A robust system to validate inputs and generate structured prompts based on style.

```python
from langchain_core.prompts import PromptTemplate

# Task 5: Input Validation Layer
def validate_inputs(audience, tone):
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    final_audience = audience if audience in valid_audiences else "beginner"
    final_tone = tone if tone in valid_tones else "formal"

    return final_audience, final_tone

# Task 6: Prompt Generator App
def generate_prompt(topic, audience, tone, style):
    v_audience, v_tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Generate interview questions about {topic} for an {audience} with a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    template_str = styles.get(style, styles["teaching"])
    prompt_temp = PromptTemplate.from_template(template_str)
    return prompt_temp.format(topic=topic, audience=v_audience, tone=v_tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test
Demonstrating 5 different inputs using a single reusable template.

```python
from langchain_core.prompts import PromptTemplate

# Task 7: Template Reusability Test
reusable_template = PromptTemplate.from_template("Concept: {concept} | Level: {level} | Action: {action}")

data_inputs = [
    {"concept": "Recursion", "level": "Junior", "action": "Debug"},
    {"concept": "Docker", "level": "DevOps", "action": "Deploy"},
    {"concept": "API", "level": "Frontend", "action": "Connect"},
    {"concept": "SQL", "level": "Analyst", "action": "Query"},
    {"concept": "Git", "level": "Student", "action": "Commit"}
]

for item in data_inputs:
    print(reusable_template.format(**item))
```
